```python
import hdf5storage
import xarray as xr
import numpy as np
```

```python
st = hdf5storage.loadmat('simtest3a_gps.mat')
```

```python
dsattrib = st['datasetAttrib'].dtype.names
datasetAttrib = dict()
for i in range(len(dsattrib)):
    if (st['datasetAttrib'][dsattrib[i]][0].size != 0):
        datasetAttrib[dsattrib[i]] = st['datasetAttrib'][dsattrib[i]][0][0][0]
    else:
        datasetAttrib[dsattrib[i]] = []
```

```python
ds = xr.Dataset(
    data_vars = dict(
        obsdata_North = (["space", "time"], st['obsData'][:,:,0]),
        obsdata_East = (["space", "time"], st['obsData'][:,:,1]),
        obsdata_Up = (["space", "time"], st['obsData'][:,:,2]),
        pntCrd_North = (["space"], st['pntCrd'][:,0]),
        pntCrd_East = (["space"], st['pntCrd'][:,1]),
        pntCrd_Up = (["space"], st['pntCrd'][:,2]),
        #pntName= (["space"], st['pntName'][:,0])
    ),
    coords = dict(
        space = range(5),
        time = st['epochDyear'][0]
    ),
    attrs=dict(datasetId = st['datasetId'], techniqueId = st['techniqueId'][0][0], datasetattrib= datasetAttrib)
)
```
